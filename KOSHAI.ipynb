```python
import pandas as pd
```

```python
#reading all Excel files
A=pd.read_excel('/content/A.xlsx')
B=pd.read_excel('/content/B.xlsx')
C=pd.read_excel('/content/C.xlsx')
D=pd.read_excel('/content/D.xlsx')
```

```python
print(A.shape)
print(B.shape)
print(C.shape)
print(D.shape)
```

```python
#Outer join on orderId

Mergeddf=pd.merge(A,B, left_on=A['A_OrderId'],right_on=B['B_OrderId'],how= 'outer')
Mergeddf
```

```python
Mergeddf1=pd.merge(C,D, left_on=['A_OrderId'],right_on=['B_OrderId'],how= 'outer')
```

```python
# fuunction for check reconciliation status
def reconciled(row):
  if pd.notnull(row['A_OrderId']) and pd.notnull(row['B_OrderId']):
        return 'reconciled'
  elif pd.notnull(row['A_OrderId']):
        return 'Present In 1st'
  elif pd.notnull(row['B_OrderId']):
        return 'Present In 2nd'
  else:
        return None
```

```python
Mergeddf['Status'] = Mergeddf.apply(reconciled, axis=1)

Mergeddf1['Status'] = Mergeddf1.apply(reconciled, axis=1)


```

```python
Mergeddf1
```

```python
##Appending both dataframes
X=Mergeddf.append(Mergeddf1).drop_duplicates(subset=['A_OrderId','B_OrderId']).reset_index(drop=True)
X.to_csv('X.csv')
```
